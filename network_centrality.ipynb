# ライブラリの読み込み

```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N, p = 8, 1/8 # ノード数とエッジがつながっている確率
G = nx.erdos_renyi_graph(N, p, seed = 314, directed = True)
# ノードの設定
node_color = "indigo"
alpha = 0.5
edgecolors = "black"
edge_color = "dimgray"
# エッジの設定
edge_color = "dimgray"
width = 2.0
arrowsize = 30

nx.draw(G, with_labels = True, pos = nx.kamada_kawai_layout(G), \
    node_size = 2000, node_color = node_color, alpha = alpha, edgecolors = edgecolors, \
    edge_color = edge_color, width = width, arrowsize = arrowsize
    )
```

# 近接中心性
4と6が若干大きい

```python
centrality = nx.closeness_centrality(G)
node_size = np.array(list(centrality.values())) * 4000
nx.draw(G, with_labels = True, pos = nx.kamada_kawai_layout(G), \
    node_size = node_size, node_color = node_color, alpha = alpha, edgecolors = edgecolors, \
    edge_color = edge_color, width = width, arrowsize = arrowsize
    )
plt.savefig("近接中心性.png")
```

# 次数中心性

```python
centrality = nx.degree_centrality(G)
node_size = np.array(list(centrality.values())) * 4000
nx.draw(G, with_labels = True, pos = nx.kamada_kawai_layout(G), \
    node_size = node_size, node_color = node_color, alpha = alpha, edgecolors = edgecolors, \
    edge_color = edge_color, width = width, arrowsize = arrowsize
    )
plt.savefig("次数中心性.png")
```

# 固有ベクトル中心性

```python
centrality = nx.eigenvector_centrality(G)
node_size = np.array(list(centrality.values())) * 4000
nx.draw(G, with_labels = True, pos = nx.kamada_kawai_layout(G), \
    node_size = node_size, node_color = node_color, alpha = alpha, edgecolors = edgecolors, \
    edge_color = edge_color, width = width, arrowsize = arrowsize
    )
plt.savefig("固有ベクトル中心性.png")
```

# ページランク

```python
centrality = nx.pagerank(G)
node_size = np.array(list(centrality.values())) * 8000
nx.draw(G, with_labels = True, pos = nx.kamada_kawai_layout(G), \
    node_size = node_size, node_color = node_color, alpha = alpha, edgecolors = edgecolors, \
    edge_color = edge_color, width = width, arrowsize = arrowsize
    )
plt.savefig("ページランク.png")
```

# 媒介中心性

```python
centrality = nx.betweenness_centrality(G)
node_size = np.array(list(centrality.values())) * 10000
nx.draw(G, with_labels = True, pos = nx.kamada_kawai_layout(G), \
    node_size = node_size, node_color = node_color, alpha = alpha, edgecolors = edgecolors, \
    edge_color = edge_color, width = width, arrowsize = arrowsize
    )
plt.savefig("媒介中心性.png")
```
